```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(0, math.pi * 2, 0.05)
y =  np.sin(x)
```

```python
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin wave')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
